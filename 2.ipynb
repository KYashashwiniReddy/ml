**Hypothesis:**  
- H0: distribution of tip for yellow = distribution of tip for green 
- H1:distribution of tip for yellow â‰  distribution of tip for green

```python
yellow_tip=df['tip'].loc[df['color']=='yellow']
green_tip=df['tip'].loc[df['color']=='green']
```

```python
yellow_tip.shape
```

```python
green_tip.shape
```

```python
# Independent T-Test
from scipy import stats
t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
if p_val < 0.05:   
    # 0.05->significance level
    print("Reject null hypothesis:The tip for yellow is diff that of green")
else:
    print("Fail to reject null hypothesis")
```

Hypothesis:

H0: Distribution of tip for cash = distribution of tip for credit card

H1: Distribution of tip for cash â‰  distribution of tip for credit card

```python
cash_tip = df['tip'].loc[df['payment'] == 'cash']
card_tip = df['tip'].loc[df['payment'] == 'credit card']
```

```python
from scipy import stats
t_stat, p_val = stats.ttest_ind(card_tip, cash_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
if p_val < 0.05:
    print("Reject null hypothesis: Payment method affects tip amount")
else:
    print("Fail to reject null hypothesis")
```

**Hypothesis:**  
- H0: distribution of tip is the same across all regions  
- H1: At least one category has a different distribution of tip

```python
Mahattan_tip=df['tip'].loc[df['pickup_borough']=='Manhattan']
Queens_tip=df['tip'].loc[df['pickup_borough']=='Queens']
Bronx_tip=df['tip'].loc[df['pickup_borough']=='Bronx']
Brooklyn_tip=df['tip'].loc[df['pickup_borough']=='Brooklyn']
```

```python
# ANOVA Test-analyse groups
f_stat, p_val = stats.f_oneway(Mahattan_tip,Queens_tip,Bronx_tip,Brooklyn_tip)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)
if p_val < 0.05:
    print("Reject null hypothesis: tip differs by pickup_borough")
else:
    print("Fail to reject null hypothesis: No significant difference by tip")

```

**Hypothesis for Tukey test (for each pair of groups):**  
- H0: distribution of group1 = distribution of group2  
- H1: distribution of group1 â‰  distribution of group2

```python
# Tukey HSD Test
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey HSD
tukey_result = pairwise_tukeyhsd(endog=df.tip, groups=df.pickup_borough, alpha=0.05)
print(tukey_result)
```
